```python
import pandas as pd
import os

# íŒŒì¼ ê²½ë¡œ ë¦¬ìŠ¤íŠ¸
file_paths = [
    r"G:\.shortcut-targets-by-id\1aXfYtUWSYS8foz14MAJMhDH7IHhG1wNZ\2ì¡°\ë°ì´í„°\clinical model\liver\clinical_data_liver.csv",
    r"G:\.shortcut-targets-by-id\1aXfYtUWSYS8foz14MAJMhDH7IHhG1wNZ\2ì¡°\ë°ì´í„°\clinical model\kidney\TCGA-KIRP_clinical_data.csv",
    r"G:\.shortcut-targets-by-id\1aXfYtUWSYS8foz14MAJMhDH7IHhG1wNZ\2ì¡°\ë°ì´í„°\clinical model\ovarian\TCGA-OV_clinical_data.csv",
    r"G:\.shortcut-targets-by-id\1aXfYtUWSYS8foz14MAJMhDH7IHhG1wNZ\2ì¡°\ë°ì´í„°\clinical model\stomach\TCGA-STAD_clinical_data.csv"
]

# íŒŒì¼ëª… ë§¤í•‘
file_names = ['LIVER (ê°„ì•”)', 'KIDNEY (ì‹ ì¥ì•”)', 'OVARIAN (ë‚œì†Œì•”)', 'STOMACH (ìœ„ì•”)']

def display_all_columns():
    print("=" * 80)
    print("ğŸ“‹ CSV íŒŒì¼ë³„ ì „ì²´ ì»¬ëŸ¼ ëª©ë¡")
    print("=" * 80)
    
    for i, file_path in enumerate(file_paths):
        try:
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                columns = list(df.columns)
                
                print(f"\nğŸ—‚ï¸  {file_names[i]}")
                print(f"ğŸ“ íŒŒì¼ëª…: {os.path.basename(file_path)}")
                print(f"ğŸ“Š ì´ ì»¬ëŸ¼ ê°œìˆ˜: {len(columns)}")
                print(f"ğŸ“ ë°ì´í„° í–‰ ê°œìˆ˜: {len(df)}")
                print("-" * 60)
                
                # ì»¬ëŸ¼ì„ ë²ˆí˜¸ì™€ í•¨ê»˜ ì¶œë ¥
                for idx, col in enumerate(columns, 1):
                    print(f"  {idx:2d}. {col}")
                
                print("-" * 60)
                
            else:
                print(f"\nâŒ {file_names[i]}")
                print(f"ğŸ“ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {os.path.basename(file_path)}")
                print(f"ğŸ” ê²½ë¡œ: {file_path}")
                
        except Exception as e:
            print(f"\nâŒ {file_names[i]}")
            print(f"ğŸ“ íŒŒì¼: {os.path.basename(file_path)}")
            print(f"ğŸš« ì˜¤ë¥˜: {str(e)}")

def display_columns_side_by_side():
    """ì»¬ëŸ¼ë“¤ì„ ë‚˜ë€íˆ ë¹„êµí•´ì„œ ë³´ì—¬ì£¼ëŠ” í•¨ìˆ˜"""
    all_data = {}
    
    print("\n" + "=" * 100)
    print("ğŸ“Š íŒŒì¼ë³„ ì»¬ëŸ¼ ë¹„êµí‘œ")
    print("=" * 100)
    
    # ê° íŒŒì¼ì˜ ì»¬ëŸ¼ ìˆ˜ì§‘
    for i, file_path in enumerate(file_paths):
        try:
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                all_data[file_names[i]] = list(df.columns)
            else:
                all_data[file_names[i]] = ["íŒŒì¼ ì—†ìŒ"]
        except Exception as e:
            all_data[file_names[i]] = [f"ì˜¤ë¥˜: {str(e)}"]
    
    # ìµœëŒ€ ì»¬ëŸ¼ ìˆ˜ ì°¾ê¸°
    max_cols = max(len(cols) for cols in all_data.values())
    
    # í—¤ë” ì¶œë ¥
    print(f"{'ìˆœë²ˆ':<4} {'LIVER':<25} {'KIDNEY':<25} {'OVARIAN':<25} {'STOMACH':<25}")
    print("-" * 104)
    
    # ì»¬ëŸ¼ë³„ë¡œ ì¶œë ¥
    for i in range(max_cols):
        row = f"{i+1:<4} "
        for file_name in file_names:
            if i < len(all_data[file_name]):
                col_name = all_data[file_name][i]
                # ì»¬ëŸ¼ëª…ì´ ë„ˆë¬´ ê¸¸ë©´ ìë¥´ê¸°
                if len(col_name) > 23:
                    col_name = col_name[:20] + "..."
                row += f"{col_name:<25} "
            else:
                row += f"{'':25} "
        print(row)

def save_columns_to_file():
    """ì»¬ëŸ¼ ì •ë³´ë¥¼ í…ìŠ¤íŠ¸ íŒŒì¼ë¡œ ì €ì¥"""
    with open("column_analysis.txt", "w", encoding="utf-8") as f:
        f.write("CSV íŒŒì¼ë³„ ì»¬ëŸ¼ ë¶„ì„ ê²°ê³¼\n")
        f.write("=" * 50 + "\n\n")
        
        for i, file_path in enumerate(file_paths):
            try:
                if os.path.exists(file_path):
                    df = pd.read_csv(file_path)
                    columns = list(df.columns)
                    
                    f.write(f"{file_names[i]}\n")
                    f.write(f"íŒŒì¼ëª…: {os.path.basename(file_path)}\n")
                    f.write(f"ì»¬ëŸ¼ ê°œìˆ˜: {len(columns)}\n")
                    f.write("ì»¬ëŸ¼ ëª©ë¡:\n")
                    
                    for idx, col in enumerate(columns, 1):
                        f.write(f"  {idx:2d}. {col}\n")
                    
                    f.write("\n" + "-" * 50 + "\n\n")
                    
            except Exception as e:
                f.write(f"{file_names[i]} - ì˜¤ë¥˜: {str(e)}\n\n")
    
    print("ğŸ“„ ê²°ê³¼ê°€ 'column_analysis.txt' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

# ì‹¤í–‰
if __name__ == "__main__":
    # 1. ê° íŒŒì¼ë³„ ìƒì„¸ ì»¬ëŸ¼ ì¶œë ¥
    display_all_columns()
    
    # 2. ì»¬ëŸ¼ ë¹„êµí‘œ ì¶œë ¥
    display_columns_side_by_side()
    
    # 3. íŒŒì¼ë¡œ ì €ì¥
    save_columns_to_file()
    
    print("\nâœ… ëª¨ë“  ë¶„ì„ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!")
```
